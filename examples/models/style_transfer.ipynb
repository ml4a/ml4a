## Run a basic style transfer


```python
from ml4a import image
from ml4a.models import style_transfer

config = {
    'content_image': image.tubingen(),
    'style_images': [image.starrynight()],
    'size': (500, 900),
    'content_weight': 0,
    'initial_iterations': 1000,
    'iterations': 500,
    'gpus': [0]
}

img0 = image.monalisa()
img = style_transfer.run(config, None)
image.display(img)
```
